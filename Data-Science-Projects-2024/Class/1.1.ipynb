### Basic API

- estimator.fit(X, [y])
    - model 을 학습
    - y는 정답 값이기 때문에 필수가 아님
- predict   :: Classification | Regression | Clustering
- transform :: Preprocessing | Dimensionality reduction | Feature selection | Feature extraction



### Classification

- 분류 :: 고객이 구매를 할지 안 할지
- Decision Tree

```python
from sklearn import tree
# tree 패키지 안에 DecisionTreeClassifier 모듈이 있음

# 데이터 설정
X = [[0, 0], [1, 1]]
Y = [0, 1]
```

```python
clf = tree.DecisionTreeClassifier()
clf
```

```python
# 데이터 학습
clf = clf.fit(X, Y)
```

```python
clf.predict([[2., 2.]])
```

```python
# 비율로 출력
clf.predict_proba([[2., 2.]])
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
X, y

# DecisionTree 알고리즘은 숫자만 학습 가능
```
